```python
import ast

def convert_genres(obj):
    res=[]
    for item in ast.literal_eval(obj):
        if item['name']:
            res.append(item['name'])
    return res
```

```python
movies['genres']=movies['genres'].apply(convert_genres)
```

```python
movies.head(10)
```

```python
movies.iloc[0].keywords
```

```python
def convert_keywords(obj):
    res=[]
    for item in ast.literal_eval(obj):
        if item['name']:
            res.append(item['name'])
    return res
```

```python
movies['keywords']=movies['keywords'].apply(convert_keywords)
```

```python
movies.head(10)
```

```python
movies.iloc[0].cast
```

```python
def convert_cast(obj):
    res=[]
    counter=0
    for item in ast.literal_eval(obj):
        if item['name']:
            if counter == 5:
                break
            else:
                res.append(item['name'])
                counter+=1
    return res
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies.head(10)
```

```python
movies.iloc[0].cast
```

```python
def convert_crew(obj):
    res=[]
    for item in ast.literal_eval(obj):
        if item['job'] == 'Director':
                res.append(item['name'])
                break
    return res
```

```python
movies['crew']=movies['crew'].apply(convert_crew)
```

```python
movies.iloc[0].overview
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies.head(10)
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head(10)
```

```python
movies['tags'] = movies['overview']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head(10)
```

```python
movies=movies[['movie_id','title','tags']]
```

```python
movies['tags']=movies['tags'].apply(lambda x:" ".join(x).lower())
```

```python
movies.head(10)
```

```python
#text vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector=cv.fit_transform(movies['tags']).toarray()
```

```python
for i, word in enumerate(cv.get_feature_names()):
    print(i, word)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
ps.stem('loving')
```

```python
def stem(text):
    res=[]
    
    for word in text.split():
        res.append(ps.stem(word))
    return " ".join(res)
```

```python
movies['tags'].apply(stem)
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity=cosine_similarity(vector)
```

```python
similarity.shape
```

```python
similarity
```

```python
def recommend(mv):
    movie_idx=movies[movies['title'] == mv].index[0]
    distances=similarity[movie_idx]
    movies_lst=sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for mv in movies_lst:
        print(mv)
        
    for mv in movies_lst:
        print(movies.iloc[mv[0]],mv[1], sep="\t\t")
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
